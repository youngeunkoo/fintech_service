```python
final_result = []
START_INDEX=1

while True:
    
    url = "http://openapi.seoul.go.kr:8088/6c735868656b79793731675676456d/xml/dreamMoney/1/5/"

    seoul_s_key = os.getenv('seoul_s_key')
    payload = dict(KEY=seoul_s_key, SERVICE='dreamMoney', START_INDEX=1 ,END_INDEX=100, TYPE='xml')


    r = requests.get(url, params = payload)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    # json으로 안하고 xml 로 받았음 

    print(r.text)
    
    lo_org_cd = [item.string for item in soup.find_all('lo_org_cd')]
    lo_org_nm = [item.string for item in soup.find_all('lo_org_nm')]
    base_str_dt = [item.string for item in soup.find_all('base_str_dt')]
    base_end_dt = [item.string for item in soup.find_all('base_end_dt')]
    lo_cnt = [item.string for item in soup.find_all('lo_cnt')]
    min_lo_ir = [item.string for item in soup.find_all('min_lo_ir')]
    max_lo_ir = [item.string for item in soup.find_all('max_lo_ir ')]
    avg_lo_ir = [item.string for item in soup.find_all('avg_lo_ir')]
    min_lo_ale_ir = [item.string for item in soup.find_all('min_lo_ale_ir')]
    max_lo_ale_ir = [item.string for item in soup.find_all('max_lo_ale_ir')]
    avg_lo_ale_ir = [item.string for item in soup.find_all('avg_lo_ale_ir')]
    
    
    result = dict(
    lo_org_cd= lo_org_cd,
    lo_org_nm=lo_org_nm,
    base_str_dt= base_str_dt,
    base_end_dt=base_end_dt,
    lo_cnt=lo_cnt,
    min_lo_ir=min_lo_ir,
    max_lo_ir= max_lo_ir,
    avg_lo_ir=avg_lo_ir,
    min_lo_ale_ir=min_lo_ale_ir,
    max_lo_ale_ir=max_lo_ale_ir,
    avg_lo_ale_ir=avg_lo_ale_ir
    )
    
    final_result.append(result)
    
    break
```

```python
final_df = pd.DataFrame()

for item in final_result:
    temp = pd.DataFrame(item)
    final_df = pd.concat([final_df, temp])

final_df

# 몇개 없기 때문에 이렇게 굳이 하지 않아도 됨
```

```python
keys = {}

for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue 
        else:
            keys.setdefault(tag.name, "")

result = {}       
for key in keys:
    for item in soup.find_all(key):
        result.setdefault(item.name, []).append(item.string)
        
        
result




result_df = pd.DataFrame(result)
result_df
```
