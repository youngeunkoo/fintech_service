# 예제 2 : 공공데이터 - 서울특별시 강남구_의류수거함 위치 현황

```python
# step 1

url = "https://api.odcloud.kr/api/15127131/v1/uddi:a9873b46-9551-407a-aff5-a3a77befb3d4"



servicekey = os.getenv('servicekey')
payload = dict(serviceKey=servicekey, page=1, perPage=100, returnType='JSON')
# 공공데이터 정보에서 파라미터 명칭 잘 확인하기

r = requests.get(url, params = payload)

print(r, url)
print(r.status_code)
print(r.text)
data = r.json()
data
```

```python
data['data']
```

```python
# step 1

cloth_bins = []
page = 1

while True:

    url = "https://api.odcloud.kr/api/15127131/v1/uddi:a9873b46-9551-407a-aff5-a3a77befb3d4"


    servicekey = os.getenv('servicekey')
    payload = dict(serviceKey=servicekey, page=page, perPage=10, returnType='JSON')
    # 공공데이터 정보에서 파라미터 명칭 잘 확인하기

    r = requests.get(url, params = payload)


    print(r.status_code)
    data = r.json()
    cloth_bins.append(data['data'])
    total_pages = data['totalCount'] // data['perPage'] + 1
    print(f"{page}/{total_pages}", end = "\r")
    
    if page < total_pages: 
        page += 1
    else: 
        break
        
        
        
        
result = pd.DataFrame()

for bins in cloth_bins[:]: 
    temp = pd.json_normalize(bins)
    result = pd.concat([result, temp])
    
result
# result.to_csv(f"naver_book_api_fintech_{query}_result.csv", encoding = "cp949")
```

# 예제 3 : 소상공인시장진흥공단 , 치킨집 데이터만 분류하기

```python
# step 1

url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"


servicekey = os.getenv('servicekey')
payload = dict(serviceKey=servicekey, divId='indsSclsCd', key='I21006' ,numOfRows = 100, pageNo =1, type='json')
# 공공데이터 정보에서 파라미터 명칭 잘 확인하기

r = requests.get(url, params = payload)

print(r, url)
print(r.status_code)
print(r.text)
data = r.json()
data
```

```python
# step 2  

chicken_data = []
pageNo = 1

while True:

    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"

    servicekey = os.getenv('servicekey')
    payload = dict(serviceKey=servicekey, divId='indsSclsCd', key='I21006' ,numOfRows = 100, pageNo =1, type='json')
    # 공공데이터 정보에서 파라미터 명칭 잘 확인하기

    r = requests.get(url, params = payload)



#     print(r.status_code)
    data = r.json()
    
    chicken_data.append(data['body']['items'][:])
    total_pages = data['body']['totalCount'] // data['body']['pageNo'] 
    print(f"{pageNo}/{total_pages}", end = "\r")
    
    if page < total_pages: 
        page += 1
    else: 
        break
        
        
        
        
result = pd.DataFrame()

for chicken in chicken_data[:]: 
    temp = pd.json_normalize(chicken)
    result = pd.concat([result, temp])
    
result
```
