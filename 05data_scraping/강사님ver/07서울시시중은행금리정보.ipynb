```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv
from bs4 import BeautifulSoup as bs
load_dotenv()
```

```python
url = f"http://openapi.seoul.go.kr:8088/{os.getenv('seoul_s_key')}/xml/dreamMoney/1/100/"
r = requests.get(url)
# print(r.status_code)
# print(r.text)
soup = bs(r.text, 'xml')
keys ={}
for item in soup.find_all('row'):
    for tag in item:
        if tag.name == None:
            continue
        else:
            keys.setdefault(tag.name, "")
result = {}
for key in keys:
    for item in soup.find_all(key):
        result.setdefault(item.name, []).append(item.string)
result_df = pd.DataFrame(result)
result_df
```

```python
with open("xml.xml", "w", encoding='utf-8') as f:
    f.write(r.text)
```

```python
import xml.etree.ElementTree as ET

# XML 파일 로드
tree = ET.parse('xml.xml')
root = tree.getroot()
print(root)
# 데이터를 저장할 빈 리스트 생성
data = []

# XML의 각 <row> 요소를 반복하며 데이터 추출
for elem in root.iter('row'):
    row = {}
    for tag in elem:
        row[tag.tag] = tag.text
    data.append(row)

# 데이터를 pandas DataFrame으로 변환
df = pd.DataFrame(data)
df
```
