```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
import time
load_dotenv()
```

```python
chicken_shops = []
page = 1
while True:
    url = "http://apis.data.go.kr/B553077/api/open/sdsc2/storeListInUpjong"
    payload = dict(serviceKey= os.getenv('servicekey'), pageNo=page,\
                  numOfRows=1000, divId='indsSclsCd', key='I21006', type='json')
    r = requests.get(url, params=payload)
    data = r.json()
    chicken_shops.append(data)
    total_pages = data['body']['totalCount'] // data['body']['numOfRows'] + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    if page < total_pages:
        page += 1
    else:
        break
    time.sleep(0.2)
```

```python
print(len(chicken_shops))
```

```python
print(len(chicken_shops[0]['body']['items']))
```

```python
total_chicken_shop_info = pd.DataFrame()
for chicken_shop in chicken_shops:
    temp = pd.json_normalize(chicken_shop['body']['items'])
    total_chicken_shop_info = pd.concat([total_chicken_shop_info, temp])
total_chicken_shop_info.to_csv("전국치킨가게정보.csv", encoding="cp949", index=False)
```
