```python
url = "http://apis.data.go.kr/B551408/rent-loan-rate-info/rate-list"
servicekey = os.getenv('servicekey')
payload = dict(serviceKey=servicekey, numOfRows=30, pageNo=1, dataType='JSON')
r = requests.get(url, params=payload)
print(r.url)
print(r.status_code)
data = r.json()
data
```

```python
result = pd.json_normalize(data['body']['items'])
result
```

```python
cloth_bins = []
page = 1

while True:
    url = "https://api.odcloud.kr/api/15127131/v1/uddi:a9873b46-9551-407a-aff5-a3a77befb3d4"
    payload = dict(serviceKey=servicekey , page=page, perPage=100, returnType="JSON")
    r = requests.get(url, params=payload)
    print(r.status_code)
    data = r.json()
    cloth_bins.append(data['data'])
    total_pages = data['totalCount'] // data['perPage'] + 1
    print(f"{page}/{total_pages}", end="\r")
    if page < total_pages:
        page += 1
    else:
        break
        
result = pd.DataFrame()
for bins in cloth_bins:
    temp = pd.json_normalize(bins)
    result = pd.concat([result, temp])
result
```
