````python
bizesId = []
bizesNm = []
brchNm = []
indsLclsCd = []
indsLclsNm = []
indsMclsCd = []
indsMclsNm = []
indsSclsCd = []
indsSclsNm = []
ksicCd = []
ksicNm = []
ctprvnCd = []
ctprvnNm = []
signguCd = []
signguNm = []
adongCd = []
adongNm = []
ldongCd = []
ldongNm = []
lnoCd = []
plotSctCd = []
plotSctNm = []
lnoMnno = []
lnoSlno = []
lnoAdr =[]
rdnmCd = []
rdnm = []
bldMnno = []
bldSlno = []
bldMngNo = []
bldNm = []
rdnmAdr = []
oldZipcd = []
newZipcd = []
dongNo = []
flrNo = []
hoNo = []
lon = []
lat = []


for page in data:
    for shop in page['body']['items']:
        bizesId.append(shop['bizesId'])
        bizesNm.append(shop['bizesNm'])
        brchNm.append(shop['brchNm'])
        indsLclsCd.append(shop['indsLclsCd'])
        indsLclsNm.append(shop['indsLclsNm'])
        indsMclsCd.append(shop['indsMclsCd'])
        indsMclsNm.append(shop['indsMclsNm'])
        indsSclsCd.append(shop['indsSclsCd'])
        indsSclsNm.append(shop['indsSclsNm'])
        ksicCd.append(shop['ksicCd'])
        ksicNm.append(shop['ksicNm'])
        ctprvnCd.append(shop['ctprvnCd'])
        ctprvnNm.append(shop['ctprvnNm'])
        signguCd.append(shop['signguCd'])
        signguNm.append(shop['signguNm'])
        adongCd.append(shop['adongCd'])
        adongNm.append(shop['adongNm'])
        ldongCd.append(shop['ldongCd'])
        ldongNm.append(shop['ldongNm'])
        lnoCd.append(shop['lnoCd'])
        plotSctCd.append(shop['plotSctCd'])
        plotSctNm.append(shop['plotSctNm'])
        lnoMnno.append(shop['lnoMnno'])
        lnoSlno.append(shop['lnoSlno'])
        lnoAdr.append(shop['lnoAdr'])
        rdnmCd.append(shop['rdnmCd'])
        rdnm.append(shop['rdnm'])
        bldMnno.append(shop['bldMnno'])
        bldSlno.append(shop['bldSlno'])
        bldMngNo.append(shop['bldMngNo'])
        bldNm.append(shop['bldNm'])
        rdnmAdr.append(shop['rdnmAdr'])
        oldZipcd.append(shop['oldZipcd'])
        newZipcd.append(shop['newZipcd'])
        dongNo.append(shop['dongNo'])
        flrNo.append(shop['flrNo'])
        hoNo.append(shop['hoNo'])
        lon.append(shop['lon'])
        lat.append(shop['lat'])
print(bizesNm[:10])
print(len(bizesNm))
print(newZipcd[:10])
print(len(newZipcd))


result = dict(bizesId=bizesId, bizesNm=bizesNm, brchNm=brchNm, indsLclsCd=indsLclsCd,
              indsLclsNm=indsLclsNm, indsMclsCd=indsMclsCd, indsMclsNm=indsMclsNm,
              indsSclsCd=indsSclsCd, indsSclsNm=indsSclsNm, ksicCd=ksicCd, ksicNm=ksicNm,
              ctprvnCd=ctprvnCd, ctprvnNm=ctprvnNm, signguCd=signguCd, signguNm=signguNm,
              adongCd=adongCd, adongNm=adongNm, ldongCd=ldongCd, ldongNm=ldongNm,
              lnoCd=lnoCd, plotSctCd=plotSctCd, plotSctNm=plotSctNm, lnoMnno=lnoMnno,
              lnoSlno=lnoSlno, lnoAdr=lnoAdr, rdnmCd=rdnmCd, rdnm=rdnm, bldMnno=bldMnno,
              bldSlno=bldSlno, bldMngNo=bldMngNo, bldNm=bldNm, rdnmAdr=rdnmAdr,
              oldZipcd=oldZipcd, newZipcd=newZipcd, dongNo=dongNo, flrNo=flrNo,
              hoNo=hoNo, lon=lon, lat=lat)
print("result: ", len(result['bizesNm']))
````

````python
%%time
chicken_shops = {}
for key in data[0]['body']['items'][0].keys():
    result = []
    for page in data:
        for shop in page['body']['items']:
            result.append(shop[key])
    chicken_shops[key] = result
print(len(chicken_shops['bizesNm']))
````

위의 식을 리스트 표현식으로 변경

````python
%%time
chicken_shops = {}
for key in data[0]['body']['items'][0].keys():
    chicken_shops[key] = [shop[key] for page in data for shop in page['body']['items']]
print(len(chicken_shops['bizesNm']))
````
