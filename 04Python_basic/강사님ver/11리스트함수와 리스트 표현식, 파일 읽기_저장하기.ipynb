## 리스트 함수
* min(): 리스트 내에서 최소값 찾아주는 함수
* max(): 리스트 내에서 최대값을 찾아주는 함수
* sum(): 리스트 내의 요소를 더해 합계를 구해주는 함수

```python
a = [38, 21, 53, 62, 19, 5, 78, 55, 42, 33]
b = [77, 65, 48, 79, 5, 4, 35, 3, 99, 105]
```

* 정렬해서 가장 앞에 있는 숫자 추출

```python
sorted(a)[-1]
```

```python
sorted(a, reverse=True)[-1]
```

```python
# 리스트 내에서 가장 큰 값을 찾는 코드
temp = b[0]
for i in b:
    if temp <= i:
        temp = i
print("최대값:", temp )
```

```python
# 리스트 내에서 가장 큰 값을 찾는 코드
temp = b[0]
for i in b:
    if temp >= i:
        temp = i
print("최소값:", temp)
```

```python
print(min(a))
print(max(a))
print(min(b))
print(max(b))
```

# 파일 읽기, 쓰기 open(), with open()

## 1) 파일에 문자열을 쓸 때는 open 함수로 파일을 열어서 파일 객체(file object)를 얻을 뒤에 write 메서드를 사용해 저장

* (1)파일 저장하기<br>
변수명 = open(파일이름, 파일모드(w:쓰기, r:읽기)<br>
변수명.write('저장할 문자열')<br>
변수명.close() * close()를 안하면 계속 열려 있음.

```python
file = open('test.txt', 'w')  # test.txt로 파일 만들기 copy con test.txt
file.write('파일 만들기 연습')  # test.txt에 .write('내용') 으로 내용 쓰기  , 내용입력
file.close()   # text.txt 파일 저장하고 닫기
```

* (2) 파일 읽기<br>

변수명 = open(파일이름, 파일모드('r'읽기)<br>
변수명2 = 변수명.read()<br>
변수명.close()

```python
file2 = open('test.txt', 'r')
my_file = file2.read()
file2.close()
print(my_file)
```

## with 문으로 파일 자동으로 닫기
* 파일을 만들려면 파일 만들기 => 내용 적기 => 파일닫기(저장)<br>
* 이때 close()를 안하면 오류 발생<br>
* 이런 실수를 줄이기 위해서는 파일을 자동으로 닫아주는 with문을 사용하면 편리

```python
with open('test_with.txt', 'w') as file3:
    file3.write('with open을 사용하면 파일을 자동으로 닫아주어 편리하다.')
```

```python
with open('test_with.txt', 'r') as file4:
    text = file4.read()
print(text)
```

# 딕셔너리 표현식 
* {} 안에 반복문을 넣어서 딕셔너리를 만드는 방법
* {key : value로 올 for문 key로 올 for문}
* {key : value for value in data key for in keys}

```python
result3 = {key : [item[key] for item in data] for key in list(data[0].keys())}
print(result3)
```

```python
result4 = {key: [item[key] for item in data] for key in list(data[0].keys())}
print(result4)       
```

* setdefault 함수를 이용해 만들기

```python
dic = {}
for datum in data:
    for k, v in datum.items():
        dic.setdefault(k, []).append(v)
print(dic)
```
