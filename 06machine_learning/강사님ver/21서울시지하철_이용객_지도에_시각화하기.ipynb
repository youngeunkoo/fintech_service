```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
scaled = mms.fit_transform(total[['people_in', 'people_out']])
scaled = pd.DataFrame(scaled, columns=['people_in', 'people_out'] )
scaled
```

```python
total['people_in'] = scaled['people_in']
total['people_out'] = scaled['people_out']
```

```python
total
```

```python
from folium.plugins import HeatMapWithTime
```

```python
heat_data = []
for timestamp, group in total.groupby('timestamp'):
    heat_map_data = []
    for _ , row in group.iterrows():
        heat_map_data.append([row['geo_latitude'], row['geo_longitude'], row['people_in']])
    heat_data.append(heat_map_data)
```

```python
heat_data[:3]
```

```python
from folium.plugins import HeatMapWithTime
```

```python
# timestamp를 시간대별로 그룹핑하여 HeatMapWithTime에 맞는 데이터 준비
heat_data = []
time_index = []  # 시간 인덱스를 기록할 리스트
for timestamp, group in total.groupby('timestamp'):
    heat_map_data = [
        [row['geo_latitude'], row['geo_longitude'], row['people_in']]
        for _, row in group.iterrows()
    ]
    heat_data.append(heat_map_data)
    time_index.append(timestamp)  # 각 시간대를 기록

# folium 지도 생성 (초기 중심 좌표 설정)
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# HeatMapWithTime 추가
HeatMapWithTime(heat_data, radius=15).add_to(m)

# 지도 저장
# m.save("heatmap_with_time.html")

# 출력할 필요가 있을 경우 지도 반환 (주피터 노트북 등에서)
# m
```
