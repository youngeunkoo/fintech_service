* 예제 : while 문으로 1-10까지 출력하기

```python
i = 1 
while i <= 10:
    print(i)
    i+= 1
```

```python
i = 1 
while i <= 10:
    i+= 1
    print(i)

#  이렇게 바꾸면 결과값 달라짐.
```

* 예제 : while 문 사용해서 1-10까지 더해서 결과를 출력하세요.

방법 1 과 방법 2는 print 위치에 따라 결과가 다르게 출력된다. 

```python
# 방법1 

i = 1
tem = 0
while i <= 10:
    tem += i
    i += 1
    print(tem)
```

```python
# 방법 2

i = 1
tem = 0
while i <= 10:
    tem += i
    i += 1
print(tem)
```

* 예제 : 1-50 까지 수에서 3의 배수이거나, 짝수인 수를 찾아서 더한 값을 출력하세요.
<br> <br>
- 1~50 반복
- 3의 배수 or 짝수
- 더하기
- 출력하기

```python
# 옳은 답변! 

i = 1
tem3 = 0

while i <= 50: 
    if i % 3 == 0 or i % 2 == 0:
        tem3 += i
    i += 1
            
print(tem3)
```

```python
# 잘못된 답변! i += 1 의 위치가 중요함

i = 1
tem3 = 0

while i <= 50: 
    if i % 3 == 0 or i % 2 == 0:
        tem3 += i
        i += 1
            
print(tem3)
```

## 반복 횟수가 정해지지 않은 경우 : 무한반복

ex) 데이터분석할 때, 크롤링 경우 <br>
네이버 도서 구매 화면, 사이트 내 역사 책이 총 몇 권 있는지 모른다. 무한반복을 돌려서 마지막 역사 책이 있을 때 까지 while 문을 돌려서 총 몇권 있는지 찾아낸다. 더이상 없을 때 break 를 걸어 멈추게 한다.

<br> 
ex) securities 리스트에는 분석할 증권사들의 종목 코드가 포함되어 있습니다.
각 증권사에 대해 Yahoo Finance API를 사용하여 2023년의 주가 데이터를 가져옵니다.
데이터가 없는 경우(data.empty == True) 해당 증권사의 데이터를 가져올 수 없다는 메시지를 출력하고 break 문을 사용하여 반복을 멈춥니다.
데이터가 있는 경우(data.empty == False) 데이터프레임에 주가 정보를 추가합니다.
이 예시에서 break 문은 데이터가 없는 증권사에 대해 데이터 수집을 중단시키는 역할을 합니다. 이후에는 분석 작업을 추가하여 각 증권사의 주가 데이터를 기반으로 필요한 분석을 수행할 수 있습니다.
<br>

* 무한반복을 하다가 숫자가 100이 되는 시점까지 숫자를 더해서 합을 구하세요.

```python
i = 1
hap = 0

while True:
    if i <= 100:
        hap += i
    
    else:
        break
    i += 1 

print(hap)
```

### * True 가 되는 경우(무한반복이 되는 경우) : 1, 정수, 실수, 문자열, list, tuple, dictionary, True
### * False 가 되는 경우 : 0, 0.0, ", "", [ ] , ( ) , { }

```python
while []:
    print("무한반복")
```

## break : 반복문을 정지시키는 명령

## continue : 특정 조건에서 코드 실행 건너뛰기 

* 예제 : 1-100까지 수 중에서 짝수면 더하고 홀수면 건너뛰기

```python
hap = 0

for i in range(1, 101):
    if i % 2 == 1:
        continue 
    else: 
        hap += i 
        print(hap)
```

## 중첩반복문

* 분침 0 ~ 59분, 시침이 1 움직임

* 시계 만들기 

```python
for si in range(0, 24):  # 시침
    print(si)
```

```python
for si in range(0, 24):     # 시침
    for bun in range(0, 60):   # 분침
        print(si, "시", bun, "분")
```

```python
for si in range(0, 24):     # 시침
    for bun in range(0, 60):   # 분침
        for cho in range(0, 60):   #초침
            print(si, "시", bun, "분", cho, "초")
```
