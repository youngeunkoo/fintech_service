

### 원본 데이터 중, 딕셔너리 자료형일 때 값이 누락된 경우 : get 사용!
<br>
파이썬에서 get 함수는 주로 딕셔너리 dictionary 자료형에서 사용됩니다. 이 함수는 특정 키(key)에 해당하는 값을 가져오는 데 사용되며, 키가 존재하지 않을 때 기본값을 반환할 수 있게 해줍니다. 이를 통해 키가 존재하지 않을 경우 발생할 수 있는 오류를 방지할 수 있습니다. <br><br>

get 함수의 기본 구조: 

dictionary.get( key, default_value )

- key: 가져오려는 값의 키입니다.
- default_value (선택 사항): 키가 사전에 없을 때 반환할 기본값입니다. 이 매개변수를 생략하면 기본값으로 None이 반환됩니다.

```python
# key 값만 출력해보기 

for item in data[:1] :
    for i in item :
        print(i)
```

```python
# 즉, 이 tour_data에서는 누락된 key 값 none으로 반환하기 위해 
# 슬라이싱으로 3번까지만 출력해보기

for item in data[:3] :
    for i in item :
        print(i, item.get(i, 'none'))
    
```

```python
for items in data[:5]:
    for i in items:
        print(i, items.get(i, 'none'))

# 만약, key 값 중 빈값이 있다면 'none' 으로 반환해라 ~
```

```python
# 이번에는 value 값만 출력해보기

for item in data[:3]:
    for i in item.values():
        print(i)
```

```python
# 방법 1 : 가장 기초적인 방법 

ForNum = []
NatNum = []
addrCd = []
gungu = []
resNm = []
rnum = []
sido = []
yyyymm = []

for item in data[:] :
    ForNum.append(item['ForNum'])
    NatNum.append(item['NatNum'])
    addrCd.append(item['addrCd'])
    gungu.append(item['gungu'])
    resNm.append(item['resNm'])
    rnum.append(item['rnum'])
    sido.append(item['sido'])
    yyyymm.append(item['yyyymm'])

result = dict(ForNum=ForNum, NatNum=NatNum, addrCd=addrCd, gungu=gungu, resNm=resNm, rnum=rnum, sido=sido, yyyymm=yyyymm)

print(result) 
```

# 딕셔너리 표현식 
* {} 안에 반복문을 넣어서 딕셔너리를 만드는 방법
* {key : value로 올 for문 key로 올 for문}
* {key : value for value in data key for in keys}

```python
result3 = {key : [item[key] for item in data] for key in list(data[0].keys())}
print(result3)
```

```python
result4 = {key: [item[key] for item in data] for key in list(data[0].keys())}
print(result4)       
```

* setdefault 함수를 이용해 만들기

```python
# 방법 4

dic = {}
for datum in data:
    for k, v in datum.items():
        dic.setdefault(k, []).append(v)
print(dic)
```
