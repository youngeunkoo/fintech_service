### 이상지출 탐지하기 1
- 평소 사용자가 지출하던 금액보다 과도하게 높은 경우
- z 스코어를 이용해 탐지
- z-score = (원래있던값 - 평균) / 표준편차

$$
z-score  =  \frac{{\text{amt} - \text{mean}}}{\text{std}}
$$


```python
train['amt_z_score'] = (train['amt'] - train['mean']) / train['std']
```

```python
train[['amt', 'mean', 'std', 'amt_z_score']]
```

```python
amt_z_over_3 = train[train['amt_z_score'] > 3]
```

# FLAML 을 이용한 AutoML 분석
- FLAML을 이용한 AutoML 분석은 머신러닝 모델을 자동으로 탐색하고 최적의 하이퍼파라미터를 찾아주는 과정을 자동화하여, 빠르고 효율적인 머신러닝 모델 구축을 가능하게 하는 방법이다. FLAML은 특히 시간과 자원을 절약하는 데 중점을 두어, 사용자가 모델과 하이퍼파라미터를 일일이 선택할 필요 없이 최적의 설정을 자동으로 찾아준다.

* FLAML : Fast and Lightweight AutoML
- 빠르고 가벼운 자동화된 머신러닝(AutoML) 라이브러리다. FLAML은 최소한의 리소스로 빠르게 고성능의 머신러닝 모델을 찾아주는 데 목적을 두고 있다. 특히 하이퍼파라미터 최적화를 간소화하고, 적은 메모리와 시간을 사용하는 것이 특징이다.
- FLAML은 빠르고 가벼운 AutoML 프레임워크로, 자동화된 머신러닝 과정을 간편하게 수행할 수 있게 도와준다. 제한된 시간과 자원 안에서 최적의 성능을 낼 수 있는 모델을 찾는 데 최적화된 도구다.

```python
#!pip install "flaml[catboost]"
```

```python
from flaml import AutoML
from sklearn.metrics import classification_report
```

```python
automl = AutoML(time_budget=600, metric='roc_auc', task='classification', log_file_name='fraud_detection.log' )

automl.fit(X_smt, y_smt)
pred = automl.predict(X_valid)
print(classification_report(y_valid, pred))
```
