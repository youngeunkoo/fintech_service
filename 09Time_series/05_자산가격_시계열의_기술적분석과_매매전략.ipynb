### 애플주식 AAPL 대한 평균 수익률 구하고 그래프 그리기

```python
data.columns
```

```python
sym = 'AAPL.O'
data = pd.DataFrame(data[sym]).dropna()
data
```

### 20일간의 애플주식 rolling 지표 구하기 

```python
window = 20 
```

```python
data['min'] = data[sym].rolling(window).min()            # 20 일 기간의 이동 최저값
data['mean'] = data[sym].rolling(window).mean()         # 20 일 기간의 이동 평균
data['std'] = data[sym].rolling(window).std()          # 20 일 기간의 이동 표준편차 
data['median'] = data[sym].rolling(window).median()   # 20 일 기간의 이동 중앙값
data['max'] = data[sym].rolling(window).max()        # 20 일 기간의 이동 최대값
data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()   # 20 일 기간의 이동 가중평균

data
```

### 가중 이동 평균(Exponentially Weighted Moving Average, EWMA)

가중 이동 평균(EWMA)은 이동 평균(Moving Average)의 한 종류로, 최근의 데이터에 더 큰 가중치를 주고, 과거 데이터에는 점점 더 작은 가중치를 주는 방식이다. 이 방식을 통해 최신 정보를 더 반영하면서도 과거 데이터를 완전히 무시하지 않는 계산 방법이다.
<br><br>
이동 평균은 특정 기간 동안의 평균을 구하는데, 보통 단순 이동 평균(SMA)은 모든 데이터를 동일한 가중치로 취급해서 평균을 계산한다. 하지만 EWMA는 시간이 지남에 따라 가중치를 다르게 줘서, 최근 데이터가 더 중요한 경우에 유용하다.
<br><br>
예를 들어 설명하면: <br>
단순 이동 평균(SMA): 5일 동안의 주가가 10, 12, 13, 15, 14라고 하면, 5일간의 평균은 (10 + 12 + 13 + 15 + 14) / 5 = 12.8로 단순하게 계산한다. 모든 날의 주가에 동일한 중요도(가중치)를 준다.
<br><br>
지수 가중 이동 평균(EWMA): 똑같은 5일 동안의 주가를 보더라도, 최근 주가가 더 중요할 수 있다. 예를 들어, 마지막 날(14)의 주가가 더 중요한 정보라고 보면, EWMA는 최신 데이터에 더 많은 가중치를 부여해서 평균을 계산한다. 과거의 주가는 가중치가 더 작아진다.
<br><br>

EWMA가 왜 유용한가? <br>
주가와 같은 데이터는 시간이 지남에 따라 변동성이 크기 때문에, 최신 데이터가 더 중요한 경우가 많다. 예를 들어, 최근 몇 주의 주가 변화가 다음 주의 주가를 예측하는 데 더 중요한 역할을 할 수 있지. EWMA는 이런 점을 반영해서 계산되는 방법이야.
<br><br>
정리하자면, EWMA는 최근 데이터에 더 많은 가중치를 부여해서 최신 변화에 더 민감하게 반응하도록 하는 평균 계산 방식이다.

## 골든크로스 (Golden Cross)
* 단기 이동평균선이 장기 이동평균선을 아래에서 위로 교차할 때
* 주가의 단기 이동평균선(보통 50일)이 장기 이동평균선(보통 200일)을 상향 돌파할 때 발생한다. 이는 주가가 상승 추세로 전환될 가능성이 높다는 신호로 해석된다. 
* 상승장의 시작을 알리는 신호
    * 단기 이동 평균선 : 보통 50일 이동 평균선을 사용
    * 장기 이동 평균선 : 보통 200일 이동 평균선을 사용

<br>

## 데드크로스 (Dead Cross)
* 단기 이동평균선이 장기 이동평균선을 위해서 아래로 교차할 때
* 반대로, 주가의 단기 이동평균선이 장기 이동평균선을 하향 돌파할 때 발생한다. 이는 주가가 하락 추세로 전환될 가능성이 높다는 신호로 해석된다.
    * 단기 이동 평균선 : 보통 50일 이동 평균선을 사용
    * 장기 이동 평균선 : 보통 200일 이동 평균선을 사용

```python
# 골든 크로스와 데드 크로스 그래프 같이 그리기 
data['positions'] = np.where(data['short_ma'] > data['long_ma'], 1, -1)  

# 1 일때는 상승장 = 매수 
# -1 일때는 하락장 = 매도

ax = data[[sym, 'short_ma', 'long_ma', 'positions']].plot(figsize=(12,6), secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
ax.set_title("AAPL Trading Window based on Technical Analysis")
plt.show()
```
